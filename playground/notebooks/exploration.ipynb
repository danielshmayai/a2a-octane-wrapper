## 1. Discover the Agent Card

````python
import asyncio, json
from a2a_client import A2AClient

async def get_card():
    async with A2AClient() as c:
        return await c.agent_card()

card = asyncio.run(get_card())
print(json.dumps(card, indent=2))
````

## 2. Send a Single Message

````python
async def ask(question):
    async with A2AClient() as c:
        return await c.send(question)

response = asyncio.run(ask("Tell me a joke"))
print(response)
````

## 3. Multi-Turn Conversation

````python
# Create a persistent client for multi-turn
client = A2AClient()
print(f"Session: {client.context_id}")

async def chat(msg):
    resp = await client.send(msg)
    print(f'User : {msg}')
    print(f'Agent: {resp}\n')
    return resp

# Run turns sequentially
async def run_conversation():
    await chat("Get defect 2110")
    await chat("What is its status?")
    await chat("Who is assigned?")
    await client._http.aclose()

asyncio.run(run_conversation())
````

## 4. Health Check

````python
async def health():
    async with A2AClient() as c:
        return await c.health()

print(asyncio.run(health()))
````
